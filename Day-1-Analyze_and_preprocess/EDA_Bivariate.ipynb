# Bivariate Data Analysis: Multivariate 
<p>Bivariate analysis involves the analysis of two variables to determine the empirical relationship between them</p>
<p>   

    Scatter Plot:

    Used to visualize the relationship between two continuous variables.
    Each point represents an observation in the dataset with the x-coordinate as one variable and the y-coordinate as the other variable.
    Helps in identifying patterns, trends, and correlations between the two variables.

Line Plot:

    Used when the variables are ordered, such as time-series data.
    Each point is connected by a line to show the relationship between the two variables over time or another ordered variable.

Hexbin Plot:

    Used to represent the relationship between two continuous variables when the data points are dense.
    Instead of plotting each point individually, hexbin plots use hexagonal bins to represent the density of points in different regions of the plot.
     
 </p>

### Mathematical example


```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset (heights and weights of individuals)
heights = np.array([160, 165, 170, 175, 180, 185])
weights = np.array([60, 65, 70, 75, 80, 85])

# Scatter plot of height against weight
plt.figure(figsize=(8, 6))
plt.scatter(heights, weights, color='blue')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset (time and temperature)
time = np.array([0, 1, 2, 3, 4, 5])
temperature = np.array([20, 22, 25, 27, 24, 22])

# Line plot of temperature over time
plt.figure(figsize=(8, 6))
plt.plot(time, temperature, marker='o', color='red')
plt.title('Temperature Over Time')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset (heights and weights of individuals)
heights = np.random.normal(170, 10, 1000)
weights = np.random.normal(70, 10, 1000)

# Hexbin plot of height against weight
plt.figure(figsize=(8, 6))
plt.hexbin(heights, weights, gridsize=20, cmap='Blues')
plt.colorbar(label='count in bin')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

### Bivariate Analysis 

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
flights=sns.load_dataset("flights")
df=pd.read_csv("titanic.csv")
```

## Using scatter plot (numerical vs numerical)

```python
tips.head(3)
```

```python
sns.scatterplot(x=tips["total_bill"],y=tips["tip"],hue=tips["sex"],style=tips["smoker"])
```

### using bar plot (numerical vs categorical)

```python
sns.barplot(x=df["Pclass"],y=df["Fare"],hue=df["Sex"])
```

### Box plot (numerical vs categorica)

```python
sns.boxplot(x=df["Age"],y=df["Sex"],hue_norm=df["Survived"])
```

### Heatmap (categorical vs categorcal)

```python
#use cross table to find frequency between two varivable 

pd.crosstab(df["Pclass"],df["Survived"])
```

```python
sns.heatmap(pd.crosstab(df["Pclass"],df["Survived"]),annot=True)
```

```python
## now use p class to find the percetages of people survived

df.groupby("Pclass")["Survived"].mean()*100
```
